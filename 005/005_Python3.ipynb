```python
import numpy as np
import csv
```

```python
dataPath = 'data.csv'
```

```python
def get_data(path):
    f = open(path, 'r')
    reader = csv.reader(f, delimiter=',')
    header = next(reader)
    data = []
    for line in reader:
      
        data.append([float(x) for x in line])
        
    return header, np.transpose(data)
```

```python
header, data = get_data(dataPath)
```
